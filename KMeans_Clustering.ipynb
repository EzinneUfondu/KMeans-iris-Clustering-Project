# Loading Dataset

```python
data=pd.read_csv('/Users/zee/Downloads/a08a1080b88344b0c8a7-0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
data.info()
```

```python
data.head()
```

# Data Preprocessing 

```python
x=data.drop('species', axis=1)
x
```

# Converting to a NumPy Array

```python
x=x.values
x
```

```python
# Training the KMeans Clustering Model
```

```python
from sklearn.cluster import KMeans
c3=KMeans(n_clusters=3)
c3.fit(x)
```

# Viewing the Cluster Centers

```python
c3.cluster_centers_
```

# Viewing the Cluster Labels

```python
c3.labels_
```

# Checking the Model's Inertia- Measure how tight or compact the clusters are from the centroid

```python
c3.inertia_
```

# Visualising the Clusters/Adding Titles and Axis Labels 

```python
plt.scatter(x[:,2], x[:,3], c=c3.labels_)
plt.title('Iris Clusters by Petal Measurements')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

# Visualised clusters with centroids

```python
plt.scatter (x[:, 2], x[:, 3], c=c3.labels_), plt.scatter (c3.cluster_centers_[:, 2], c3.cluster_centers_[:, 3], color='red')

plt.title('Iris Clusters with Centroids')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
inertias=[]
for K in range(1,11):
    cK=KMeans(n_clusters=K)
    cK.fit(x)
    inertias.append(cK.inertia_)
```

# Plotting the Elbow Curve

```python
plt.plot(range(1,11),inertias)
plt.title('Elbow Method - Optimal Number of Clusters')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
```

```python
data['cluster'] = c3.labels_
print(data.groupby(['species', 'cluster']).size())
```
